```python
# Explore the relationship between the feature and target column using a count plot of categorical columns and a scatter plot of numerical columns
```

```python
# count plot for categorical columns : Sex, Smoker - charges
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=data, x='smoker')
plt.title("count of Smoker and Non-Smoker")
plt.xlabel("smoker")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=data, x='region', hue = "smoker")
plt.title("count of Smoker and Non-Smoker")
plt.xlabel("smoker")
plt.ylabel("count")
plt.show()
```

```python
Observations : In all regions, there are more non-smokers than smokers.
```

```python
# scatter plot of numerical columns
```

```python
sns.scatterplot(data=data, x='age', y='charges')
```

```python
sns.scatterplot(data=data, x='bmi', y='charges')
```

Observations :
Age : There is noticeable upward trend-change tend to increase with age especially with older individuals
BMI : A slight upward trend can be seen , but it is not very strong. some high charges aprear at higher bmi levels.
Children : No clear correlation between number of children and charges.

```python
# Perform data visualization using plots of feature vs feature
```

```python
fig, axs = plt.subplots(2,3, figsize=(18, 10))

sns.scatterplot(data=data, x='age', y='bmi', hue='smoker', ax=axs[0,0])
axs[0,0].set_title("Age vs BMI")

sns.scatterplot(data=data, x='age', y='children', hue='smoker', ax=axs[0,1])
axs[0,1].set_title("Age vs children")

sns.scatterplot(data=data, x='age', y='charges', hue='smoker', ax=axs[0,2])
axs[0,2].set_title("Age vs charges")

sns.scatterplot(data=data, x='bmi', y='children', hue='smoker', ax=axs[1,0])
axs[1,0].set_title("BMI vs Children")

sns.scatterplot(data=data, x='bmi', y='charges', hue='smoker', ax=axs[1,1])
axs[1,1].set_title("BMI vs Charges")

sns.scatterplot(data=data, x='children', y='charges', hue='smoker', ax=axs[1,2])
axs[1,2].set_title("Children vs Charges")

plt.tight_layout()
```

Observations :
Age vs Charges: Charges increase with age, especially for smokers. A clear upward trend is visible for smokers, indicating a strong age-related risk factor.

BMI vs Charges: Some smokers with high BMI are charged significantly more, but the correlation is weaker than with age.

Age vs BMI / Children: No strong relationshipBMI and number of children appear fairly independent of age.

Children vs Charges: No clear pattern; the number of children has little visible effect on charges.

BMI vs Children: Again, little to no trend here.

```python
# Check if the number of premium charges for smokers or non-smokers is increasing as they are aging
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=data, x='age', y='charges', hue = "smoker", estimator='mean')
plt.title("Avg insurance charges vs Age by Smoking status")
plt.xlabel("Age")
plt.ylabel("Average Charges")
plt.show()
```

Observations :
Smokers : Charges rise steeply with age , espesically after age 30. This suggests an accelerating cost with age for smokers.

Non-smokers : Charges also increase with age , but at much slower rate and more gradually.

This confirms a strong age-related rise in premium cost, particularly pronounced for smokers.
